```python
import requests
import re
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup





```

```python

url="https://en.wikipedia.org/wiki/Python_(programming_language)"
def get_html(url):
    rep=requests.get(url)
    return  rep.content
def extract_title(html):
    soup=BeautifulSoup(html,'html.parser')
    title=soup.find('title').text
    return title
def extract_content(html):
    soup = BeautifulSoup(html, 'html.parser')
    content = {}
    titres = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
    for titre in titres:
        
        heading_text = titre.text
        #print(heading_text)
        next_elem = titre.find_next_sibling()
        paragraph = ''
        while next_elem and next_elem.name not in ['h1', 'h2', 'h3', 'h4', 'h5','h6']:
            if next_elem.name == 'p':
                paragraph += '\n' + next_elem.text
            next_elem = next_elem.find_next_sibling()
        content[heading_text] = paragraph
    return content

def get_links(html):
    soup = BeautifulSoup(html, 'html.parser')
    links = []
    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.startswith('/wiki/') and ':' not in href:
            links.append('https://en.wikipedia.org' + href)
    return links

def get_data(url):
    html = get_html(url)
    title = extract_title(html)
    content = extract_content(html)
    links = get_links(html)
    data = {'title': title, 'content': content, 'links': links}
    return data
               
       
dic={}
dic=get_data(url)

print(dic)
```
